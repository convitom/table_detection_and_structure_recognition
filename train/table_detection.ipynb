```python
from ultralytics import YOLO

model = YOLO("/content/drive/MyDrive/runs/table_detect/yolov8_table/weights/last.pt")
model.train(
    resume=False,

    data="/content/dataset_table/data.yaml",
    project="/content/drive/MyDrive/runs/table_detect",
    name="yolov8_table",
    device=0,

    epochs=20,          # hoặc tổng epoch mới
    imgsz=640,
    batch=8,
    workers=4,

    optimizer="AdamW",
    lr0=1e-3,
    cos_lr=True,

    mosaic=0.3,
    shear=2.0,
    translate=0.1,
    scale=0.5,

    hsv_h=0.0,
    hsv_s=0.0,
    hsv_v=0.3,

    fliplr=0.0,
    flipud=0.0,

    patience=20,
    save_period=1,
    verbose=True,
)
```
